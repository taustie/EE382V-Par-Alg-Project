```python
import numpy
%matplotlib inline
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams.update({'font.size': 22})
```

```python
# Read random input points

with open('input.txt') as f:
    lines = f.readlines()
    x = [line.split(',')[0] for line in lines]
    y = [line.split(',')[1] for line in lines]

input_x = [int(value.split('(')[1]) for value in x]
input_y = [int(value.split(')')[0]) for value in y]
input_x = numpy.array(input_x)
input_y = numpy.array(input_y)

# for index, x in enumerate(input_x):
#     print(index, ':', input_x[index], input_y[index])
```

```python
# Read Convex Hull output points

with open('output.txt') as f:
    lines = f.readlines()
    x = [line.split(',')[0] for line in lines]
    y = [line.split(',')[1] for line in lines]

output_x = [int(value.split('(')[1]) for value in x]
output_y = [int(value.split(')')[0]) for value in y]
output_x = numpy.array(output_x)
output_y = numpy.array(output_y)

# for index, x in enumerate(output_x):
#     print(index, ':', output_x[index], output_y[index])
```

```python
plt.title('Convex Hull')
# plt.plot(input_x, input_y, 'o', color='black');
plt.scatter(input_x, input_y, s=10, color ='black', label="input")
plt.plot(output_x, output_y, '-o', color='red', label="on hull")
# plt.legend(loc="upper left")
plt.grid(True)
```
